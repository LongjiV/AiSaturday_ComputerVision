```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Define Cartoonizer Function**

```python
#read in image
image = cv2.imread(r"C:\Users\CCECC\Pictures\COLORFUL-NIGHT.jpg")

def cartoonize(img):
    #convert to RGB
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    #convert image to grayscale
    gray = cv2.cvtColor(img_RGB, cv2.COLOR_BGR2GRAY)

    #apply median filter to image
    gray = cv2.medianBlur(gray, 5)

    #create cartoon color
    color = cv2.bilateralFilter(img_RGB, 9, 300, 300)

    #detect edges
    edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 9)
    edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)

    #cartoon effect
    cartoon = cv2.bitwise_and(color, edges)

    fig = plt.figure(figsize=(20,20))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    fig1 = ax1.imshow(img_RGB)
    fig2 = ax2.imshow(cartoon)

    return fig1 , fig2
```

```python
cartoonize(image)
```
